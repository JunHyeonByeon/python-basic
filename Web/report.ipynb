# 우리동네 맛집 찾기

```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
from urllib.parse import urljoin
import pandas as pd
import re
from tqdm import tqdm_notebook
import time
import numpy as np
```

```python
url_base = 'https://www.diningcode.com/'
url_sub ='list.php?query=%EB%8C%80%EC%A0%84%20%EC%9C%A0%EC%84%B1%EA%B5%AC%20%EC%8B%A0%EC%84%B1%EB%8F%99'
url = url_base + url_sub

html = urlopen(url)
soup = BeautifulSoup(html, 'html.parser')
```

```python
rest_list = soup.find_all(class_='blink')
rest_list
```

```python
len(rest_list)
```

```python
favor = soup.find_all(class_='favor-review')
favor
```

```python
rest = []
menu = []
point = []
url_add = []
```

```python
for list1 in rest_list:
    rest.append(list1.find(class_='btxt').get_text()[3:])
    menu.append(list1.find(class_='stxt').get_text())
    url_add.append(urljoin(url_base, list1['href']))
```

```python
for list2 in favor:
    point.append(list2.find(class_='point').get_text())
```

```python
df = pd.DataFrame({'가게명' : rest, '메뉴' : menu, '평점' : point, 'URL' : url_add})
df
```

```python
tel = []
addr = []
for n in tqdm_notebook(df.index):
    html = urlopen(df['URL'][n])
    soup_tmp = BeautifulSoup(html, 'html.parser')
    
    tel.append(soup_tmp.find(class_='tel').get_text())
    addr.append(soup_tmp.find(class_='locat').get_text())
```

```python
df.head()
```

```python
del df['URL']
df.head()
```

```python
df['전화번호'] = tel
df['주소'] = addr

df.set_index('가게명', inplace=True)
df
```
